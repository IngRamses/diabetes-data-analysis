An√°lisis Exploratorio de Datos (EDA)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")

# Distribuci√≥n de la variable dependiente
sns.countplot(x='Diabetes_012', data=df)
plt.title("Distribuci√≥n de clases de diabetes")
plt.show()

# Histogramas de edad y BMI
df['BMI'].hist(bins=30, edgecolor='black')
plt.title("Distribuci√≥n del IMC")
plt.show()
```

An√°lisis Estad√≠stico

```python
from scipy.stats import chi2_contingency, ttest_ind

# Prueba chi-cuadrado para HighBP
contingency = pd.crosstab(df['HighBP'], df['Diabetes_012'])
chi2, p, _, _ = chi2_contingency(contingency)
print(f"Chi-cuadrado HighBP vs Diabetes: p = {p:.4f}")

# t-test entre clase 0 y 2 para BMI
df_binario = df[df['Diabetes_012'] != 1]
test0 = df_binario[df_binario['Diabetes_012'] == 0]['BMI']
test2 = df_binario[df_binario['Diabetes_012'] == 2]['BMI']
t_stat, p_val = ttest_ind(test0, test2)
print(f"t-test IMC: p = {p_val:.4f}")
```

Modelo Predictivo: Regresi√≥n Log√≠stica Binaria

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

# Preparar datos
df_bin = df[df['Diabetes_012'] != 1]
df_bin['Diabetes'] = df_bin['Diabetes_012'].replace({0: 0, 2: 1})
X = df_bin.drop(columns=['Diabetes_012', 'Diabetes'])
y = df_bin['Diabetes']

# Dividir dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train_const = sm.add_constant(X_train)

# Modelo
model = sm.Logit(y_train, X_train_const).fit()
print(model.summary())
```

Evaluaci√≥n del Modelo

```python
# Predicciones y m√©tricas
X_test_const = sm.add_constant(X_test)
y_pred_prob = model.predict(X_test_const)
y_pred_class = (y_pred_prob >= 0.5).astype(int)

print(confusion_matrix(y_test, y_pred_class))
print(classification_report(y_test, y_pred_class))

# Curva ROC
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("Curva ROC")
plt.legend()
plt.show()
```

Validaci√≥n del Modelo (VIF y Odds Ratios)

```python
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices

# Calcular VIF
y_vif, X_vif = dmatrices("Diabetes ~ " + " + ".join(X.columns), data=df_bin, return_type='dataframe')
vif_data = pd.DataFrame()
vif_data['Variable'] = X_vif.columns
vif_data['VIF'] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
print(vif_data.sort_values("VIF", ascending=False))

# Odd Ratios
odds_ratios = pd.DataFrame({
    'Variable': model.params.index,
    'Odds Ratio': model.params.apply(lambda x: round(np.exp(x), 2)),
    'p-value': model.pvalues
})
print(odds_ratios)
```
